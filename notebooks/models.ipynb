```python
cylindrical_img = Image.open('/Users/umaobbani/Documents/DSR_ai-ev-battery-shape-classification_portfolio/ai-ev-battery-shape-classification/data/processed/train/images/cylindrical/1_1_1_cyl.jpg').convert("RGB")
cylindrical_img 
```

```python
cylindrical_img_transformed = transform(cylindrical_img).unsqueeze(0)
cylindrical_img_transformed = cylindrical_img_transformed.to(device)
cylindrical_img_transformed
```

```python
import torch.nn.functional as F

resnet_model.eval()
with torch.inference_mode():
    logits = resnet_model(cylindrical_img_transformed.view(1,3,224,224))
    probs = F.softmax(logits , dim=1)
    predicted_label_idx = torch.argmax(probs).item()
    predicted_label=class_labels[predicted_label_idx]
    predicted_prob = probs.flatten()[predicted_label_idx]
    print(f"label = {predicted_label}, predicted_prob = {predicted_prob}")
```
