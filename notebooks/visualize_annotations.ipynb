```python
import fiftyone as fo
from pathlib import Path

BASE_DIR = Path("/Users/umaobbani/Documents/DSR_ai-ev-battery-shape-classification_portfolio/ai-ev-battery-shape-classification/data/augumented")
SPLITS = ["train", "val", "test"]
CLASSES = ["cylindrical", "prismatic", "pouch"]

dataset = fo.Dataset(name="ev_battery_shape_classification")

for split in SPLITS:
    for cls in CLASSES:
        img_dir = BASE_DIR / split / "images" / cls

        for img_path in img_dir.glob("*.jpg"):
            sample = fo.Sample(
                filepath=str(img_path),
                ground_truth=fo.Classification(label=cls),
                tags=[split]  # important for filtering
            )
            dataset.add_sample(sample)

print(dataset)
```

```python
session = fo.launch_app(dataset)
```
